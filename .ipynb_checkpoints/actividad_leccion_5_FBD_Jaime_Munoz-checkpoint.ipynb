```python
%matplotlib inline
# Importación de librerias
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Importacion Librerias MAchine Learning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomTreesEmbedding
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import PolynomialFeatures
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#Leyendo los datos y pasandolos a dataframe
train_df = pd.read_csv('train.csv', index_col='PassengerId')
test_df = pd.read_csv('test.csv',index_col='PassengerId')
# He decidido hacer todos los cambios en un dataframe para no duplicar posteriormente
# Además IMPORTANTE, puede que si no hacemos esto, no nos percatemos de datos faltantes en ambos dataframe
combine = pd.concat([train_df,test_df])
combine.head()
```

```python
# Visualizacion general del dataframe. Con ello nos hacemos una idea de todo el conjunto de datos
profile = ProfileReport(combine, title="Profiling Report")
profile
```

```python
# Quitamos la columna Cabin por no interesarnos ya que faltan muchos datos
combine.drop('Cabin', axis=1, inplace=True)
# Quitamos la columna Ticket por no interesarnos, ya que no aporta nada.
combine.drop('Ticket', axis=1, inplace=True)
```

```python
# Ahora vamos a ver que hacemos con la columna de Embarked
# Comprobamos los valores nulos que tiene
combine.Embarked.isnull().sum()
```

```python
# Comprobamos los valores NO nulos que tiene
combine.Embarked.value_counts()
```

```python
# Rellenamos los datos faltantes con el más probable
combine['Embarked'] = combine['Embarked'].fillna('S')
combine.Embarked.value_counts()
```

```python
# Creamos la columna 'Title' de la columna Name, ya que el nombre no nos interesa, pero sí el título,
# extrayendo lo que se encuentra antes del punto
combine['Title'] = combine.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
combine.drop('Name', axis=1, inplace=True)
```

```python
# Vemos como ha quedado
# Comprobamos los valores nulos que tiene
combine.Title.isnull().sum()
```

```python
# Comprobamos los valores NO nulos que tiene
combine.Title.value_counts()
```

```python
# Vamos a agrupar los titulos que significan lo mismo o por lo menos edad semejante

combine['Title'] = combine['Title'].replace('Don', 'Mr')
combine['Title'] = combine['Title'].replace('Sir', 'Mr')
combine['Title'] = combine['Title'].replace('Col', 'Mr')
combine['Title'] = combine['Title'].replace('Major', 'Mr')
combine['Title'] = combine['Title'].replace('Capt', 'Mr')
combine['Title'] = combine['Title'].replace('Dr', 'Mr')
combine['Title'] = combine['Title'].replace('Rev', 'Mr')
combine['Title'] = combine['Title'].replace('Countess', 'Mr')
combine['Title'] = combine['Title'].replace('Jonkheer', 'Mr')

combine['Title'] = combine['Title'].replace('Mlle', 'Miss')
combine['Title'] = combine['Title'].replace('Dona', 'Mrs')
combine['Title'] = combine['Title'].replace('Lady', 'Mrs')
combine['Title'] = combine['Title'].replace('Mme', 'Mrs')
combine['Title'] = combine['Title'].replace('Ms', 'Miss')
combine.Title.value_counts()
```

```python
# Creamos una columna que es family size sacada de otras dos columnas
combine['FamilySize'] = combine['SibSp'] + combine['Parch'] + 1
```

```python
# Falta un dato en la columna Fare
combine.Fare.isnull().sum()
```

```python
combine.Fare.describe()
```

```python
# Observamos la fila que es nula y vemos que es de Pclass 3
combine[combine["Fare"].isnull()]
```

```python
combine[combine["Pclass"]==3]['Fare'].describe()
```

```python
# La media de estos billeres es 13,3 y la mediana 8,05 usaremos la mediana
combine['Fare'] = combine['Fare'].fillna(combine["Pclass"].median())
```

```python
combine['is_Alone'] = (combine['FamilySize']==1)
combine['normal_Family'] = ((combine['FamilySize']>1) & (combine['FamilySize']<5))
combine['big_Family'] = ((combine['FamilySize']>4) & (combine['FamilySize']<8))
combine['very_big_Family'] = (combine['FamilySize']>7)
```

```python
# Ahora vamos a estudiar Fare y su distribución
g = sns.distplot(combine["Fare"], color="m", label="Skewness : %.2f"%(combine["Fare"].skew()))
g = g.legend(loc="best")
```

```python
# Esa distribución no suele venir bien a los algoritmos de predicción, prefiriendo una distribución normal
# Para ello aplicaaremos la función logarítmica a todo el set
# Apply log to Fare to reduce skewness distribution
combine["Fare"] = combine["Fare"].map(lambda i: np.log(i) if i > 0 else 0)
g = sns.distplot(combine["Fare"], color="b", label="Skewness : %.2f"%(combine["Fare"].skew()))
g = g.legend(loc="best")
```

Obteniendo así una mejor distribución

```python
# Eliminamos SibSp y Parch 
combine.drop('SibSp', axis=1, inplace=True)
combine.drop('Parch', axis=1, inplace=True)
combine.drop('FamilySize', axis=1, inplace=True)
```

```python
# Ahora transformamos algunos de los datos categóricos y meto igualmente a los numericos enteros menores de 10
combine = pd.get_dummies(combine,columns=['Sex','is_Alone','normal_Family','big_Family','very_big_Family'],drop_first=True)
# después arreglaremos el resto
```

```python
combine.head()
```

```python
# Escalado de los datos de Fare
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
combine[['Fare']] = scaler.fit_transform(combine[['Fare']])
```

```python
combine.head()
```

```python
combine.columns
```

```python
# Vamos a rellenar los datos faltantes de edad, 
# Lo rellenaremos atendiendo a la mediana similar a Title, Pclass, very_big_family
# recogemos los indices que no tienen Age
index_NaN_age = list(combine["Age"][combine["Age"].isnull()].index)

for i in index_NaN_age :
    age_med = combine["Age"].median()
    age_pred = combine["Age"][((combine['Title'] == combine.iloc[i-1]["Title"]) & (combine['Pclass'] == combine.iloc[i-1]["Pclass"]) & (combine['very_big_Family_True'] == combine.iloc[i-1]["very_big_Family_True"]))].median()
    if not np.isnan(age_pred) :
        combine['Age'].iloc[i-1] = age_pred
    else :
        combine['Age'].iloc[i-1] = age_med
```

```python
combine.Age.isnull().sum()
```

```python
# Y ahora escalamos la edad igualmente
combine[['Age']] = scaler.fit_transform(combine[['Age']])
```

```python
combine = pd.get_dummies(combine,columns=['Pclass','Embarked','Title'],drop_first=True)
```

```python
# hacemos de nuevo un estudio de los datos
profile = ProfileReport(combine, title="Profiling Report")
profile
```

```python
combine.columns
```

```python
# Una vez depurados los datos y listos para entrenar
# dividimos el dataset en train y test
my_train_df = combine[combine['Survived'].notnull()]
my_test_df = combine[combine['Survived'].isnull()]
X_train=my_train_df[['Age', 'Fare', 'Sex_male', 'is_Alone_True',
       'normal_Family_True', 'big_Family_True', 'very_big_Family_True',
       'Pclass_2', 'Pclass_3', 'Embarked_Q', 'Embarked_S', 'Title_Miss',
       'Title_Mr', 'Title_Mrs']]                                
y_train=my_train_df["Survived"]
X_test = my_test_df[['Age', 'Fare', 'Sex_male', 'is_Alone_True',
       'normal_Family_True', 'big_Family_True', 'very_big_Family_True',
       'Pclass_2', 'Pclass_3', 'Embarked_Q', 'Embarked_S', 'Title_Miss',
       'Title_Mr', 'Title_Mrs']] 
```

```python
# Ahora del grupo de train,
# voy a hacer una subdivisión para comprobar algoritmos
my_X_train, my_X_test, my_y_train, my_y_test = train_test_split(X_train,y_train,random_state=10)
```

```python
# KneighborsClassifier
clf = KNeighborsClassifier()
clf.fit(my_X_train,my_y_train)
y_pred=clf.predict(my_X_test)
print('Coeficiente de determinación del conjunto de entrenamiento 2: %.4f' % clf.score(my_X_train,
                                                                         my_y_train))
print('Coeficiente de determinación del conjunto de test 2: %.4f' % clf.score(my_X_test,
                                                                    my_y_test))
```

```python
# DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(my_X_train,my_y_train)
y_pred=clf.predict(my_X_test)
print('Coeficiente de determinación del conjunto de entrenamiento 2: %.4f' % clf.score(my_X_train,
                                                                         my_y_train))
print('Coeficiente de determinación del conjunto de test 2: %.4f' % clf.score(my_X_test,
                                                                    my_y_test))
```

```python
### RANDOM FOREST
clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=19, random_state=42)
clf.fit(my_X_train,my_y_train)
y_pred = clf.predict(my_X_test)
print('Coeficiente de determinación del conjunto de entrenamiento 2: %.4f' % clf.score(my_X_train,
                                                                         my_y_train))
print('Coeficiente de determinación del conjunto de test 2: %.4f' % clf.score(my_X_test,
                                                                    my_y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(random_state = 0)
clf.fit(my_X_train, my_y_train)
y_pred = clf.predict(my_X_test)
print('Coeficiente de determinación del conjunto de entrenamiento 2: %.4f' % clf.score(my_X_train,
                                                                         my_y_train))
print('Coeficiente de determinación del conjunto de test 2: %.4f' % clf.score(my_X_test,
                                                                    my_y_test))

```

```python
# Elejimos el algoritmo de RandomForestClassifier y lo entrenamos con todo el data set
clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=19, random_state=42)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print('Coeficiente de determinación del conjunto de entrenamiento Random Forest: %.4f' % clf.score(X_train,y_train))
```

```python
# Guardamos el archvio
X_test['Survived'] = y_pred.astype(int)
X_test['Survived'].to_csv("Submission.csv")
```
